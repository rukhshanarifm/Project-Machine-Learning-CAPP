# 1. Data Conversion
   #### Flatten out json datafiles into csv files, and turn tsv into csv, and them merge these datasets

```python
import pandas as pd
import numpy as np
import json
```

```python
KEPT_COLUMNS = ['id','description', 'probe_timestamp','created_at', 'lang', 'protected', 'verified', 'geo_enabled', 
                'default_profile', 'followers_count', 'friends_count', 'listed_count', 'favourites_count', 'statuses_count','source']
def data_conversion(path_tweet, path_bot, source):
,    '''
    convert json to csv file, tsv to csv file, and merge them based on user.id
    '''
    #convert json to csv file
    with open(path_tweet) as f:
        tweet_json = json.load(f)
    tweets_df = pd.json_normalize(tweet_json)
    tweets_df.rename(columns={'created_at':'probe_timestamp'}, inplace=True)
    #remove the prefix "user."
    for column in tweets_df.columns.tolist():
        if column.startswith("user."):
            newname = column.strip("user.")
            tweets_df.rename(columns={column:newname}, inplace=True)
    #other fixes for column names (for the purpose of data consolidation later)
    if 'user_id' in tweets_df:
        tweets_df.rename(columns={'user_id':'id'}, inplace=True)
    if 'user_created_at' in tweets_df:
        tweets_df.rename(columns={"user_created_at":"created_at"}, inplace=True)
    if "tatuses_count" in tweets_df:
        tweets_df.rename(columns={"tatuses_count":"statuses_count"}, inplace=True)
    if "default_profil" in tweets_df:
        tweets_df.rename(columns={"default_profil":"default_profile"}, inplace=True)
    tweets_df = tweets_df[KEPT_COLUMNS]
    
    #convert tsv to csv file
    columns = ["id","bot"]
    bot_df = pd.read_csv(path_bot, names = columns, sep = "\t", header = None)
    # join the two datasets together by "user.id"
    merged_df =bot_df.merge(tweets_df,how = "inner", left_on='id', right_on='id')
    merged_df["source"] = source
    return merged_df
```
